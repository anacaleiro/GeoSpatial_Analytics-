## 2.1. Quick visualization to undertsand the Bus stops <a class="anchor" id="bus-viz"></a>

```python
# Not including the "Margem sul"
lisbon_stops = stops[(stops["stop_lat"] > 38.69) & 
                     (stops["stop_lon"] > -9.5) & 
                     (stops["stop_lon"] < -9.05)]
```

```python
# Ploting the map
fig = px.scatter_map(
    lisbon_stops,
    lat="stop_lat",
    lon="stop_lon",
    hover_name="stop_name",
    hover_data=["stop_id"],
    zoom=10,
    center={"lat": 38.74, "lon": -9.14},
    title="Carris Stops in Lisbon"
)

fig.update_traces(marker=dict(color="darkgreen", size=8, opacity=0.175))

fig.show()
```
